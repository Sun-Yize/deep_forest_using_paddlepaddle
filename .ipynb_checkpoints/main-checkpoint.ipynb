## **深度森林原理**

深度森林（deep forest）是周志华提出的一个集成的森林模型，是传统的森林的模型在广度和深度上的一种集成。

在深度森林中，主要包括多粒度扫描（Multi-Grained Scanning）和级联森林结构（Cascade Forest Structure）两部分。多粒度扫描通过使用不同尺度的窗口，分别对原有特征进行进一步提取；而级联森林结构则参考了深度神经网络中的神经元结构，每一层中的决策树类似于神经网络中的神经元，通过随机森林和完全随机森林的集成，构建成最终的级联森林结构。

<center><img src='https://ai-studio-static-online.cdn.bcebos.com/00ff5121616649c3b3b6f1464ed10fb920bd720087c4449f9fa02749ee0bb7d6' width=80%></center>


相比于传统的随机森林，深度森林在模型结构上具有组织性，且深度森林在深度上的优势，使其具有较强的学习能力；对比深度神经网络，深度森林所需超参数较少，更易于训练，相比于普通的深度神经网络具有较强的鲁棒性。总的来说，深度森林在当前已深度神经网络为主的时代，提供了一种新的思路，如果能解决模型的内存问题，则具有较强的潜力。

### **级联森林结构 Cascade Forest Structure**

在深度神经网络中，网络学习的过程主要依赖于逐层处理的结构。受到这种网络结构的启发，GC Forest采用了一个级联森林结构，如下图所示，其中每个级别的级联接收由其前一级处理的信息特征，并将其处理结果输出到每层的随机森林当中，即集合的集合。 在这里使用了不同类型的随机森林来增强多样性。

<center><img src='https://ai-studio-static-online.cdn.bcebos.com/7b96f6464d92406a86b8a8bedeb66008c7533cc54ba541aebf91487698e72d02' width=50%></center>


在原始论文中，作者提出以下内容，作为级联森林的每层结构：

+ 每层使用两个随机森林和两个完全随机森林。

+ 每个完全随机森林包含500棵决策树，这些决策树每次分裂的特征是随机选取的，并且每个叶子节点仅包含相同的类别。
+ 同理，每个随机森林也包含500棵决策树，决策树每次分裂的特征由Gini系数决定，选取最佳Gini系数作为拆分点。




每个森林在计算完成后，都会产生对类分布的估计，计算出叶节点的不同类别的占比，形成n维向量（n为类别数）。然后将每个森林产生的n维向量与原始数据进行拼接，作为下一层的训练特征，输入到下一个级联层级。

<center><img src='https://ai-studio-static-online.cdn.bcebos.com/bc36ca7c26084154b0d78d8d142709b6b54c188d3777435dbc09aa664cce5852' width=50%></center>

假设我们有3个类，每层有4个森林，初始向量维度为100维。每层计算完成后，4个森林中的每一个都会产生一个3维向量。因此，下一个级联级联将接收112（= 3×4+100）维的向量值。

### **多粒度扫描 Multi-Grained Scanning**

深度神经网络在处理特征关系上是强大的。由深度神经网络启发，作者在级联森林结构前添加了多粒度扫描。多粒度扫描是在原有数据特征上，用多个大小不同的滑窗在原始数据上进行滑动取值，也就是采样。

<center><img src='https://ai-studio-static-online.cdn.bcebos.com/42d3a7fbeeb04a0ca68f4d7ad65bb4ab3b98f24dbd804481b69f473fb4372150' width=50%></center>

以上图为例，假设原始数据有400维，首先使用100维大小的窗口。如果原始数据为一维向量，则窗口沿原始数据逐个进行滑动，一共生成了301个特征向量。如果原始数据是高维向量，比如20×20的图片，则由一个10×10窗口进行扫描，将产生121维的特征向量。将全部样本进行扫描，并将扫描的数据分别用随机森林与完全随机森林进行训练（类似于级联森林中每层的结构），最终生成的数据再进行拼接，得到最终的训练数据。


## **深度森林复现**

接下来我们使用python进行深度森林的复现，复现的过程中，**数据的计算与存储均基于paddle的tensor结构，没有使用numpy库，仅依赖于纯python+paddle。**

复现的过程按照顺序，依次完成以下几步：

+ 决策树结构

+ 随机森林
+ 单层的级联森林结构
+ 多粒度扫描
+ 深度森林

### **决策树结构**

按照基本的二叉树结构进行定义。同时定义calc_predict_value函数，用于计算叶子节点类别函数，以便于后期预测使用；以及describe_tree函数，便于打印出树的整体结构。

```python
# 导入飞桨 paddle
import paddle

class Tree(object):
    # 决策树基本结构
    def __init__(self):
        self.split_feature = None
        self.split_value = None
        self.leaf_value = None
        self.tree_left = None
        self.tree_right = None

    # 计算叶子节点的类别，用递归方法
    def calc_predict_value(self, dataset):
        if self.leaf_value is not None:
            return self.leaf_value
        elif dataset[self.split_feature] <= self.split_value:
            return self.tree_left.calc_predict_value(dataset)
        else:
            return self.tree_right.calc_predict_value(dataset)

    # 打印出树的结构，方便预览
    def describe_tree(self):
        if not self.tree_left and not self.tree_right:
            leaf_info = "{leaf_value:" + str(self.leaf_value) + "}"
            return leaf_info
        left_info = self.tree_left.describe_tree()
        right_info = self.tree_right.describe_tree()
        tree_structure = "{split_feature:" + str(self.split_feature) + \
                         ",split_value:" + str(self.split_value) + \
                         ",left_tree:" + left_info + \
                         ",right_tree:" + right_info + "}"
        return tree_structure
```

### **随机森林算法**

下面实现最基本的随机森林算法，与常规的随机森林算法略有不同的地方，在于其返回值为每个类别预测的概率，返回值为n维向量（n为类别数）。同时，提供使用的总特征数量作为超参数，用于分辨随机森林与完全随机森林。

以下为代码定义的随机森林中，超参数具体含义：

+ n_estimators：决策树数量
+ max_depth：决策树最大深度
+ min_samples_split：节点分裂所需的最小样本数量
+ min_samples_leaf：叶子节点最少样本数量
+ min_split_gain：分裂所需的最小增益
+ col_sample：列采样设置，选取的总特征数量
+ subsample：行采样占总样本数比例
+ label_name：类别名称

```python
class RandomForestClassifier(object):  
    # 输入各个超参数
    def __init__(self,
                 n_estimators=10,
                 max_depth=-1,
                 min_samples_split=2,
                 min_samples_leaf=1,
                 min_split_gain=0.0,
                 col_sample="sqrt",
                 subsample=0.8,
                 label_name=None):
        self.n_estimators = n_estimators
        self.max_depth = max_depth if max_depth != -1 else float('inf')
        self.min_samples_split = min_samples_split
        self.min_samples_leaf = min_samples_leaf
        self.min_split_gain = min_split_gain
        self.col_sample = col_sample
        self.subsample = subsample
        self.trees = None
        self.feature_importances_ = dict()
        self.result_class = label_name

    # 训练函数
    def fit(self, dataset, targets):
        # 设置总特征数量
        if self.col_sample == "sqrt":
            self.col_sample = int(dataset.shape[0] ** 0.5)
        else:
            self.col_sample = dataset.shape[0]
        self.trees = []
        # 生成所需决策树
        for _ in range(self.n_estimators):
            tree = self.build_trees(dataset, targets)
            self.trees.append(tree)
    
    # 决策树数据处理
    def build_trees(self, dataset, targets):
        # 选取要用的特征
        col_index = paddle.randperm(dataset.shape[1])[:self.col_sample]
        # paddle的随机抽样函数，有放回抽样，返回抽样的index
        rs = paddle.io.RandomSampler(data_source=dataset,
                                     replacement=True,
                                     num_samples=int(self.subsample*dataset.shape[0]))
        # paddle按index选取数据
        rs_index = paddle.to_tensor([temp for temp in rs])
        dataset_new = paddle.index_select(dataset, rs_index)
        targets_new = paddle.index_select(targets, rs_index)
        # 构建单棵决策树
        tree = self.build_single_tree(dataset_new, targets_new, col_index, depth=0)
        # print(tree.describe_tree())
        return tree
    
    # 构建单棵决策树
    def build_single_tree(self, dataset, targets, col_index, depth):
        # 判断是否满足退出条件
        if paddle.unique(targets).shape[0] <= 1 or dataset.shape[0] <= self.min_samples_split:
            tree = Tree()
            tree.leaf_value = self.calc_leaf_value(targets, self.result_class)
            return tree
        # 如果未到最大层数
        if depth < self.max_depth:
            best_split_feature, best_split_value, best_split_gain = self.choose_best_feature(dataset, targets, col_index)
            left_dataset, right_dataset, left_targets, right_targets = \
                self.split_dataset(dataset, targets, best_split_feature, best_split_value)
            tree = Tree()
            if left_dataset.shape[0] <= self.min_samples_leaf or \
                    right_dataset.shape[0] <= self.min_samples_leaf or \
                    best_split_gain <= self.min_split_gain:
                tree.leaf_value = self.calc_leaf_value(targets, self.result_class)
                return tree
            # 递归构建左右子树
            else:
                self.feature_importances_[best_split_feature] = \
                    self.feature_importances_.get(best_split_feature, 0) + 1
                tree.split_feature = best_split_feature
                tree.split_value = best_split_value
                tree.tree_left = self.build_single_tree(left_dataset, left_targets, col_index, depth + 1)
                tree.tree_right = self.build_single_tree(right_dataset, right_targets, col_index, depth + 1)
                return tree
        # 到最大层数
        else:
            tree = Tree()
            tree.leaf_value = self.calc_leaf_value(targets, self.result_class)
            return tree
    
    # 选取最优特征
    def choose_best_feature(self, dataset, targets, col_index):
        best_split_gain = 1
        best_split_feature = None
        best_split_value = None
        # 逐个特征计算gini系数
        for feature in col_index:
            uni = paddle.sort(paddle.unique(dataset[:, feature]))
            if uni.shape[0] <= 100:
                unique_values = uni
            else:
                unique_values = paddle.linspace(uni[0], uni[-1], 100, 'float32')[1:-2]
            # 逐个分裂点计算gini系数
            for split_value in unique_values:
                left_targets = paddle.index_select(targets, paddle.nonzero(dataset[:, feature] <= split_value))
                right_select = paddle.nonzero(dataset[:, feature] > split_value)
                if len(right_select) == 0:
                    right_targets = paddle.to_tensor([])
                else:
                    right_targets = paddle.index_select(targets, right_select)
                split_gain = self.calc_gini(left_targets, right_targets)
                # 产生最优gini系数
                if split_gain < best_split_gain:
                    best_split_feature = feature
                    best_split_value = split_value
                    best_split_gain = split_gain
        return best_split_feature, best_split_value, best_split_gain
    
    # 计算gini系数
    def calc_gini(self, left_targets, right_targets):
        split_gain = 0
        for targets in [left_targets, right_targets]:
            gini = 1
            label_counts = self.count(targets)
            for key in label_counts:
                prob = label_counts[key] * 1.0 / len(targets)
                gini -= prob ** 2
            split_gain += len(targets) * 1.0 / (len(left_targets) + len(right_targets)) * gini
        return split_gain
    
    # 计算叶子中各类数量
    def calc_leaf_value(self, targets, classes):
        label_counts = self.count(targets)
        sum_temp = sum(label_counts.values())
        result = [label_counts[x] / sum_temp if x in label_counts else 0 for x in classes]
        return result
    
    # 进行预测
    def predict(self, dataset):
        res = paddle.zeros(shape=paddle.to_tensor([dataset.shape[0], len(self.result_class)]))
        for i, row in enumerate(dataset):
            pred_list = []
            for tree in self.trees:
                pred_list.append(tree.calc_predict_value(row))
            res[i, :] = np.mean(pred_list, axis=0)
        return res
    
    # 静态方法，计算tensor中每类元素数量，返回字典
    @staticmethod
    def count(tensor):
        count_dict = {}
        for key in tensor.numpy():
            count_dict[key] = count_dict.get(key, 0) + 1
        return count_dict

    # 静态方法，利用paddle的index_select和nonzero进行数据集条件划分
    @staticmethod
    def split_dataset(dataset, targets, sf, sv):
        lda = paddle.index_select(dataset, paddle.nonzero(dataset[:, sf] <= sv))
        lta = paddle.index_select(targets, paddle.nonzero(dataset[:, sf] <= sv))
        rda = paddle.index_select(dataset, paddle.nonzero(dataset[:, sf] > sv))
        rta = paddle.index_select(targets, paddle.nonzero(dataset[:, sf] > sv))
        return lda, rda, lta, rta
```

### **单层的级联森林**

我们先定义单层的级联森林，然后再将每层进行连接。以下为单层级联森林所需超参数：

+ num_forests：每层随机森林的数量（偶数个）
+ n_estimators：每个随机森林与完全随机森林中，决策树的数量
+ max_depth：决策树最大深度
+ label_name：类别名称

```python
class Layer:
    # 输入各个超参数
    def __init__(self, n_estimators, num_forests, label_name, max_depth=30):
        self.num_forests = num_forests
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.label_name = label_name
        self.model = []
    
    # 定义训练函数
    def train(self, dataset, targets, dataset_new=None):
        result_final = dataset
        # 逐个森林进行遍历
        for forest_index in range(self.num_forests):
            # 如果为偶数，则定义随机森林
            if forest_index % 2 == 0:
                clf = RandomForestClassifier(n_estimators=self.n_estimators,
                                             max_depth=self.max_depth,
                                             label_name=self.label_name)
                clf.fit(dataset, targets)
                result = clf.predict(dataset)
                # 判断是否进行了多粒度扫描，并进行数据拼接
                if dataset_new is None:
                    result_final = paddle.concat([result_final, result], axis=1)
                else:
                    result_final = paddle.concat([result_final, result], axis=1)
            # 如果为奇数，则定义完全随机森林
            else:
                clf = RandomForestClassifier(n_estimators=self.n_estimators,
                                             max_depth=self.max_depth,
                                             col_sample=1,
                                             label_name=self.label_name)
                clf.fit(dataset, targets)
                result = clf.predict(dataset)
                # 判断是否进行了多粒度扫描，并进行数据拼接
                if dataset_new is None:
                    result_final = paddle.concat([result_final, result], axis=1)
                else:
                    result_final = paddle.concat([result_final, result], axis=1)
            self.model.append(clf)
        # 返回下一层所需数据
        return result_final
    
    # 定义预测函数
    def predict(self, dataset):
        # 定义paddle空tensor数组，存放结果
        predict_prob = paddle.zeros([dataset.shape[0], len(self.label_name) * self.num_forests])
        # 逐个森林遍历
        for f_in, clf in enumerate(self.model):
            predict_prob[:, f_in * len(self.label_name):(f_in + 1) * len(self.label_name)] = clf.predict(dataset)
        return predict_prob
```

### **多粒度扫描**

接下来定义多粒度扫描结构，主要分为初始滑窗选取，随机森林训练，数据拼接三个部分。

滑窗选取由window_slicing函数进行实现，随机森林训练和数据拼接则在train函数中进行实现。

以下为多粒度扫描所需超参数：

+ windows_size：滑动窗口大小，为列表格式

+ n_estimators：决策树数量

+ max_depth：决策树最大层数

+ label_name：类别名称

```python
class Window:
    # 输入各个超参数
    def __init__(self, windows_size, n_estimators, max_depth, label_name):
        self.windows_size = windows_size
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.label_name = label_name
        self.model = []
    
    # 定义训练函数
    def train(self, dataset, targets):
        # 逐个窗口大小进行遍历
        for window in self.windows_size:
            # 滑动窗口剪切
            pre_data = self.window_slicing(dataset, window)
            # 处理label数据
            x_data = paddle.reshape(pre_data, [dataset.shape[0] * (dataset.shape[1] - window + 1), window])
            num_w = dataset.shape[1] - window + 1
            y_data = paddle.ones([targets.shape[0] * num_w])
            for i, num in enumerate(targets):
                y_data[num_w * i:num_w * (i + 1)] = y_data[num_w * i:num_w * (i + 1)] * num
            # 保存模型列表
            model_slice = []
            # 开始训练
            print("_____training_window_rf_____")
            # 定义随机森林
            clf1 = RandomForestClassifier(n_estimators=self.n_estimators,
                                          max_depth=self.max_depth,
                                          label_name=self.label_name)
            clf1.fit(x_data, y_data)
            # 定义完全随机森林
            clf2 = RandomForestClassifier(n_estimators=self.n_estimators,
                                          max_depth=self.max_depth,
                                          col_sample=1,
                                          label_name=self.label_name)
            clf2.fit(x_data, y_data)
            # 保存模型
            model_slice.append(clf1)
            model_slice.append(clf2)
            self.model.append(model_slice)
    
    # 定义滑窗选取函数
    def window_slicing(self, dataset, window):
        pre_dataset = paddle.zeros([dataset.shape[0], dataset.shape[1] - window + 1, window])
        # 窗口遍历
        for i in range(dataset.shape[1] - window + 1):
            data_slice = dataset[:, i:i + window]
            pre_dataset[:, i, :] = data_slice
        return pre_dataset
    
    # 定义预测函数
    def predict(self, dataset):
        total_data = []
        # 逐个窗口大小进行遍历
        for i, window in enumerate(self.windows_size):
            # 滑动窗口剪切
            pre_data = self.window_slicing(dataset, window)
            x_data = paddle.reshape(pre_data, [dataset.shape[0] * (dataset.shape[1] - window + 1), window])
            num_w = dataset.shape[1] - window + 1
            model_slice = self.model[i]
            # 随机森林模型预测，生成新数据并拼接
            result1 = model_slice[0].predict(x_data)
            result1 = paddle.reshape(result1, [dataset.shape[0], num_w * len(self.label_name)])
            result2 = model_slice[1].predict(x_data)
            result2 = paddle.reshape(result2, [dataset.shape[0], num_w * len(self.label_name)])
            result_final = paddle.concat([result1, result2], axis=1)
            total_data.append(result_final)
        # 返回级联森林所需训练数据
        return total_data
```

### **深度森林**

最后实现深度森林算法，将以上各个部分进行串联，连接多粒度扫描与级联森林部分，并形成最终的深度森林。

以下为深度森林所需超参数：

+ windows_size：滑动窗口大小，为列表格式
+ n_scanning_trees：多粒度扫描所需决策树数量
+ n_estimators：级联森林所需决策树数量
+ num_forests：级联森林每层随机森林数量
+ max_layer：深度森林层数
+ max_depth：级联森林决策树最大深度
+ min_samples_leaf：叶子节点最少样本数量

```python
class GCForest:
    # 输入各个超参数
    def __init__(self,
                 windows_size=None,
                 n_scanning_trees=None,
                 n_estimators=100,
                 num_forests=4,
                 max_layer=2,
                 max_depth=31,
                 min_samples_leaf=1):
        self.windows_size = windows_size
        self.n_scanning_trees = n_scanning_trees
        self.n_estimators = n_estimators
        self.num_forests = num_forests
        self.max_depth = max_depth
        self.max_layer = max_layer
        self.min_samples_leaf = min_samples_leaf
        self.model = []
    
    # 定义训练函数
    def train(self, dataset, targets):
        self.label_name = paddle.sort(targets.unique()).numpy()
        layer_index = 0
        # 如果未定义窗口，则不进行多粒度扫描
        if self.windows_size is None:
            # 级联森林训练
            while layer_index < self.max_layer:
                print("_____layer"+str(layer_index+1)+"_construct_____")
                layer = Layer(n_estimators=self.n_estimators,
                              num_forests=self.num_forests,
                              max_depth=self.max_depth,
                              label_name=self.label_name)
                dataset = layer.train(dataset, targets)
                layer_index = layer_index + 1
                self.model.append(layer)
        # 如果定义窗口，先进行多粒度扫描
        else:
            print("_____window_slicing_____")
            # 多粒度扫描
            self.win = Window(n_estimators=self.n_scanning_trees,
                              windows_size=self.windows_size,
                              max_depth=self.max_depth,
                              label_name=self.label_name)
            self.win.train(dataset, targets)
            dataset = self.win.predict(dataset)
            use_dataset = dataset[0]
            # 级联森林训练
            while layer_index < self.max_layer:
                print("_____layer"+str(layer_index+1)+"_construct_____")
                layer = Layer(n_estimators=self.n_estimators,
                              num_forests=self.num_forests,
                              max_depth=self.max_depth,
                              label_name=self.label_name)
                use_dataset = layer.train(use_dataset, targets, dataset[(layer_index + 1) % len(self.windows_size)])
                layer_index = layer_index + 1
                self.model.append(layer)
    
    # 返回最终预测结果中，概率最大的类别
    def con_predict(self, predict):
        predict = predict.t()
        predict = predict.reshape([self.num_forests, 3, self.test_sample_num])
        result = paddle.argmax(paddle.sum(predict, axis=0).t(), axis=1)
        return result
    
    # 定义预测函数
    def predict(self, dataset):
        self.test_sample_num = dataset.shape[0]
        # 如果未定义窗口，则不进行多粒度扫描
        if self.windows_size is None:
            for layer in self.model:
                layer_result = layer.predict(dataset)
                dataset = paddle.concat([dataset, layer_result], axis=1)
        # 如果定义窗口，先进行多粒度扫描，再进行预测
        else:
            dataset = self.win.predict(dataset)
            use_dataset = dataset[0]
            for i, layer in enumerate(self.model):
                layer_result = layer.predict(use_dataset)
                use_dataset = paddle.concat([dataset[(i + 1) % len(self.windows_size)], layer_result], axis=1)
        result = self.con_predict(layer_result)
        return [self.label_name[i] for i in result.numpy()]
```

### **低差异度数列**

我们在对数据集进行抽样的过程中，使用低差异度抽样方法，可以减小误差，更接近总体分布。

有关低差异度的文献可以参考：有关低差异度的文献可以参考：[Subsampling Bias and The Best-Discrepancy Systematic Cross Validation](https://arxiv.org/abs/1907.02437)

以下是低差异度的python代码实现：

```python
class BDS:
    # 定义超越数，这里选取e作为超越数
    def __init__(self, digit):
        self.digit = digit
        self.trans = 2.71828182845904523536028747135266249775

    # 生成低差异度数列
    def get_BDS(self, x_value):
        product = x_value * self.trans
        product_decimal = round(product - int(product), self.digit)
        return float(str(product_decimal))

    # 按照低差异度数列处理数据集
    def process(self, dataset):
        x_value = dataset.values[:, :-1]
        # 正则化并降维
        x_value = preprocessing.MinMaxScaler().fit_transform(x_value)
        svd = TruncatedSVD(1)
        sort_s = svd.fit_transform(x_value)
        dataset['sort_s'] = sort_s
        dataset.sort_values(by='sort_s', inplace=True)
        bds_col = [self.get_BDS(j) for j in range(1, len(dataset) + 1)]
        dataset['BDS'] = bds_col
        dataset.sort_values(by='BDS', inplace=True)
        del dataset['BDS'], dataset['sort_s'], x_value
        return dataset
```


### **数据集分类测试**

接下来我们对数据集进行遍历测试，主要分为以下几个步骤：

+ 统计数据集的基本信息

+ 对数据集进行低差异度抽样

+ 将数据集按照k折划分

+ 分别训练深度森林和随机森林

+ 计算f1 score, precision score, accuracy score

+ 将全部数据集的计算结果可视化

以下是代码实现：

```python
# 逐个数据集进行遍历
for i1, filename in enumerate(file_names):
    print(str(i1+1) + ' nth file, filename: ' + filename)
    data = pd.read_csv(datasets_path + "/" + filename, delim_whitespace=True, header=None)
    # 划分数据集与训练集
    x_train = data.values[:, :-1]
    y_train = data.values[:, -1].astype(int)
    # 计算数据集基本信息
    classes, counts = np.unique(y_train, return_counts=True)
    balance[filename.split('.')[0]] = list(counts / y_train.size)
    # 定义k折交叉
    kf = KFold(n_splits=5, random_state=10, shuffle=True)
    accuracy_1, accuracy_2, f1_score1, f1_score2, precision1, precision2, time_cost = [[] for x in range(7)]
    start_time = time.time()
    # k折交叉验证
    for i2, (train, test) in enumerate(kf.split(x_train)):
        # 定义深度森林
        model1 = CascadeForestClassifier(
            n_estimators=2,
            n_tolerant_rounds=2,
            n_trees=100,
            max_layers=20,
            random_state=10,
            verbose=-1
        )
        model1.fit(x_train[train], y_train[train])
        y_pred1 = model1.predict(x_train[test])
        # 计算f1 score, precision score, accuracy score
        f1_1 = f1_score(y_train[test], y_pred1, average='weighted')
        precision_1 = precision_score(y_train[test], y_pred1, average='weighted')
        score_acc1 = accuracy_score(y_train[test], y_pred1)
        # 保存变量
        accuracy_1.append(score_acc1)
        f1_score1.append(f1_1)
        precision1.append(precision_1)

        # 定义随机森林
        model2 = RandomForestClassifier(
            n_estimators=100,
            min_samples_split=2,
            min_samples_leaf=1,
            random_state=0
        )
        model2.fit(x_train[train], y_train[train])
        y_pred2 = model2.predict(x_train[test])
        # 计算f1 score, precision score, accuracy score
        precision_2 = precision_score(y_train[test], y_pred2, average='weighted')
        score_acc2 = accuracy_score(y_train[test], y_pred2)
        f1_2 = f1_score(y_train[test], y_pred2, average='weighted')
        # 保存变量
        accuracy_2.append(score_acc2)
        f1_score2.append(f1_2)
        precision2.append(precision_2)
    # 将计算数据汇总至dataframe
    end_time = time.time()
    summary = summary.append(pd.DataFrame({
        'dataSet': [filename.split('.')[0]],
        'accuracy_df': [np.mean(accuracy_1)],
        'accuracy_rf': [np.mean(accuracy_2)],
        'f1_df': [np.mean(f1_score1)],
        'f1_rf': [np.mean(f1_score2)],
        'precision_df': [np.mean(precision1)],
        'precision_rf': [np.mean(precision2)],
        'classes_type': [classes.size],
        'data_std': [np.std(balance[filename.split('.')[0]])],
        'time_cost': end_time-start_time
    }), ignore_index=True)
    # print(summary.iloc[i1, :], '\n')
    # 删除本轮计算变量
    del data, model1, model2, y_pred2, y_pred1
    del x_train
    del y_train

# 将特征数量和数据数量补充到表格
summary['feature_num'] = feature_num
summary['data_num'] = row_num

# 如为了代码运行效率，则使用下两行代码读取预先训练结果
summary.to_csv('records.csv')
summary = pd.read_csv("records.csv", index_col=0)
summary.iloc[:, :-2].round(2)
```

### **假设检验**

要在某个指标上对比深度森林和随机森林模型的好坏，我们可以直接进行比较，同时为了使比较结果更具说服力，我们可以使用统计检验的方法，即将两个模型分别跑n个数据集，使用t检验比较这两组n个样本的均值、使用F检验比较这两组n个样本的方差，均值越大、方差越小说明在当前指标上该模型更好更稳定（这里说的指标是正向指标）。那么下面我们以具体代码来进行说明。

```python
# 模型一实验结果
accuracy_df=summary['accuracy_df']
# 模型二实验结果
accuracy_rf=summary['accuracy_rf']
accuracy_df=np.array(accuracy_df)
accuracy_rf=np.array(accuracy_rf)
test = ttest_rel(accuracy_df, accuracy_rf)

print('t检验结果：p值')
print(test[1])

# 计算组内样本方差
var1 = np.var(accuracy_df, ddof=1)
var2 = np.var(accuracy_rf, ddof=1)
# 计算统计量F
F = var1 / var2
# 计算自由度
df1 = len(accuracy_df) - 1
df2 = len(accuracy_rf) - 1
# 计算p值
p_value = 1 - 2 * abs(0.5 - f.cdf(F, df1, df2))
print('F检验结果：p值')
print(p_value)
```

可以看到t检验的p值小于0.05，说明拒绝原假设，即两个模型在该指标上取值的均值有差异；F检验的p值大于0.05，说明接受原假设，即两个模型结果的稳定程度一致。

通过上述比较，我们可以得出结论：从整体而言，深度森林模型优于随机森林模型。

+ 数据集数量
  + 从饼状图的占比来看，可以很明显的发现深度森林相比于随机森林来说，需要更多的数据量来进行训练。较小的数据量不能使深度森林尽快收敛。对于小于1000个数据的数据集来说，随机森林会是较好的选择。
  + 深度森林因为具有更为复杂的结构，可以处理数据量较大的数据集。

综上，深度森林在维度较高，数据量较大的数据集来说拥有较好的表现。当数据维度大于10维，数据量大于1000个时，可以优先考虑使用深度森林提高正确率。而在较小维度上或较少数据集上，虽然正确率略逊于随机森林，但仍具有较好的表现，可以作为随机森林的替补方法。
